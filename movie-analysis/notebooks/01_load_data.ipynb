```python
import pandas as pd
```

```python
# read the data from csv file
df = pd.read_csv("../data/raw/imdb_raw.csv")
```

```python
# get details about the data like each column and it's datatype
df.info()
```

```python
# Let's view statistical summaries of our dataset using the describe() method.

# Describe only numeric columns (default behavior)
print("===== NUMERIC SUMMARY (Default: Only Numeric Columns) =====")
print(df.describe())

print("\n") 

# Describe object (categorical/text) columns
print("===== CATEGORICAL SUMMARY (Using include='object') =====")
print(df.describe(include='object'))
```
